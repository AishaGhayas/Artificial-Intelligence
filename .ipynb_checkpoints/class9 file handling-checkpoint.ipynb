```python
open(URL,mode)  physical address of file, mode = r,r+, w, w+, a, a+
                                   |  r    r+     w     w+     a     a+
-----------------------------------|-----------------------------------
read                               |  +    +            +            +
write                              |       +      +     +      +     +
write after seek(pointer position) |       +      +     +
create                             |              +     +      +     +
truncate (replace data)            |              +     +
position at start                  |  +    +      +     +
position at end                    |                           +

create -->> if file not present then create
append -->> insert value at end does not disturb existing data. only mode with pointer position end
x -->> always create new file and update
w+,r+,a+ -->>if already present then update it
open()
```

with block keeps connectivity untill u are within with block. leave with block connectivity ends

- read
- readline
- readlines

```python
with open("a.txt") as abc:  #renames file a as abc
    content = abc.read()
    print(type(content))
    print(content)
    
    
#batch size is the size of chunk of data provided to read or load for efficient memory usage
```

```python
with open("a.txt") as abc: 
    content = abc.readline() #reads only one line
    print(type(content))
    print(content)
```

```python
with open("a.txt") as abc: 
    content = abc.readlines() #return list
    print(type(content))
    print(content)
```

```python
with open("a.txt") as abc:
    content = abc.readlines()[:3]
    print(type(content))
    print(content)
```

```python
with open("a.txt") as abc:
    abc.write("hello")
    
#default mode is r(read mode) write is not allowed
```

```python
with open("abc2.txt", "w") as ab:  # w -->> can write, create but not read
    ab.write("hello")
    content = ab.readlines()
    print(content)
#file has been created and hello has been written but can't read
```

```python
with open("abc2.txt", "w") as ab: 
    ab.write("hi friends") # truncate updates data and del previous
```

never use w mode for real world databases

```python
with open("abc3.txt", "a") as a3:
    a3.write("hello") #appends data as many times as it rus without changing line so u can also use "\nhello"
```

read, write, create if not present -->> a+ mode

```python
with open("abc4.txt", "a+") as a: #cursor is always at end when data is appended
    a.write("hello\n")
    c = a.readlines() #it reads after cursor as cursor is at end it gives empty list
    print(c)
```

```python
with open("abc4.txt", "a+") as a: 
    a.write("hello\n")
    a.seek(0) #brings cursor at start
    c = a.readlines() 
    print(c)
```

```python
#create file first manually and write 1,2,3
with open("abc5.txt", "r+") as f:
    a = list(reversed(f.readlines()))
    b = list(reversed(["A\n","B\n","C\n"]))
    c = a + b    #["3\n","2\n","1\n","C\n","B\n","A\n"]
    d = "".join(c)
    f.seek(0)
    f.write(d)
    print(d)
```
