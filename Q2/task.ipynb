```python
s = b.sum()
bb = pd.DataFrame(b)
bb["percent"] = bb.Status / s *100
bb
```

```python
bb.percent.plot.bar()
```

```python
dfbluex["time classes"] = pd.cut(dfbluex["Delivery Time"],4,precision=0)
dfbluex=dfbluex.sort_values("time classes")
```

```python
a = dfbluex.groupby(["Company","time classes"])[["Delivery Time"]].agg("count")
a["percentage delivery"] = a["Delivery Time"]/91 * 100
a
```
