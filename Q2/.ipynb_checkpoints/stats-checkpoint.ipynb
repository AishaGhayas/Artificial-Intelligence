Q2-12: frequency distribution

divide the following data into classes of 5 and 11 equal intervals

```python
l = [15, 32, 45, 46,42, 39,68,47,18,31,48,49,56,52,39,48,69,61,44,42,38,52,55,58,62,58,48,56,58,48,47,52,37,\
     64,29,55,38,29,62,49,69,18,61,55,49]
len(l)
```

```python
l.sort() #ascending
l
```

```python
s = pd.Series(l)
c = pd.cut(s,5).value_counts()
c
```

```python
c = pd.DataFrame(c,columns=["frequency"])
c
```

```python
c["relative frequency"] = c.frequency/45
print(c.sum(axis = 0))
c
```

```python
b = pd.cut(s,11,precision=0).value_counts()
b = pd.DataFrame(b, columns=["frequency"])
b["relative frequency"] = b.frequency/45
print(b.sum(axis=0))
b
```

Q2-16

```python
df["SAT"] = [1330,1340,1330,1310,1160,1100,1100,1250,1180,1180,1170,1150,1010,950,940,860,1070,1040,960,940]
df
```

```python
exp_sat = {2.0: 860,
           2.1: 880,
           2.2: 900,
           2.3: 910,
           2.4: 920,
           2.5: 940,
           2.6: 960,
           2.7: 980,
           2.8: 1000,
           2.9: 1020,
           3.0: 1040,
           3.1: 1070,
           3.2: 1100,
           3.3: 1130,
           3.4: 1160,
           3.5: 1190,
           3.6: 1220,
           3.7: 1250,
           3.8: 1280,
           3.9: 1310,
           4.0: 1340}
```

```python
df["expected SAT score"] = df.school.map(exp_sat)
df
```

```python
df["spread"] = df.college - df.school
df["SAT differential"] = df.SAT - df["expected SAT score"]
df.sort_values("spread", ascending=False)
```

```python
plt.scatter(df.spread,df["SAT differential"])
```

sat differential is increasing as spread(difference between college & school gpa) increases
