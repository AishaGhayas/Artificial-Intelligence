# HDF5

with HDF format u can filter data while loading therefore, used for very large datasets that dont fit in memory and small sections can be read and write efficiently without loading whole data

```python
frame = pd.DataFrame({'a': np.random.randn(100)})
store = pd.HDFStore('mydata.h5') #new database is created and stored in store
store['obj1'] = frame   # stored in object format
store['obj1_col'] = frame['a'] #a col or whole df can be stored in HDFstore
print(store) # hdf store
```

```python
store.obj1   #store["obj1"]
```

second way to store and retrieve 

```python
# store['obj2'] = frame  or
store.put("obj2", frame, format="table") # key(where to store), value, format table helps in filtering data while retrieval
# or
frame.to_hdf('mydata.h5', 'obj3', format='table') #1 line for all above lines

#retrieve
#store.obj1   #store["obj1"]
store.select("obj2", where=["index >= 10 and index <= 15"])
```

```python
store.close()
pd.read_hdf("mydata.h5", "obj3", where = ["index < 5 "])
```

# series data
* making series from:
* list
* tuple
* dict
* 1d array

```python
d1 = {"A":1,
     "B":2,
     "C":3,
     "D":4}
s3 = pd.Series(d1) # Dictionary
s3
```

```python
s3 = pd.Series(np.array([1,"aisha","A1", 1222 ])) # 1D array
s3
```

```python
s2 = pd.Series((1,2,3,4)) # Tuple
s2 = pd.Series((1,2,3,4), index=['A','B','C','D']) #s2.index = ['A','B','C','D']
s2 = pd.Series((1,2,3,4), index=['A','B','C','D'], name="CNIC") # s2.name = "CNIC"
s2.index.name="ID"
print(s2)
```
