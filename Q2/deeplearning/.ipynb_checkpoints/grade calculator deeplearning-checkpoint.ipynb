```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame(np.random.randint(1,33, size=(10000,5)), columns=["math", "phy","bio", "chem", "eng"])
df2 = pd.DataFrame(np.random.randint(34,40, size=(10000,5)), columns=["math", "phy","bio", "chem", "eng"])
df3 = pd.DataFrame(np.random.randint(41,50, size=(10000,5)), columns=["math", "phy","bio", "chem", "eng"])
df4 = pd.DataFrame(np.random.randint(51,60, size=(10000,5)), columns=["math", "phy","bio", "chem", "eng"])
df5 = pd.DataFrame(np.random.randint(61,70, size=(10000,5)), columns=["math", "phy","bio", "chem", "eng"])
df6 = pd.DataFrame(np.random.randint(71,80, size=(10000,5)), columns=["math", "phy","bio", "chem", "eng"])
df7 = pd.DataFrame(np.random.randint(81,100, size=(10000,5)), columns=["math", "phy","bio", "chem", "eng"])

df = pd.concat([df1,df2,df3,df4,df5,df6,df7])
df.head()
```

```python
df["obtained"] = df.loc[:,"math":"eng"].apply(sum,axis=1)
df["total"] = 500
df["percentage"] = df.obtained / df.total * 100
df.head()
```

# Get Grade with Symbolic AI, Classical Programming, Tranditional Programming, Rules Based programming
- Rules Always defined by programmer

```python
def grade(x):
    if x>= 80:
        return "A+"
    elif x>= 70:
        return "A"
    elif x>= 60:
        return "B"
    elif x>= 50:
        return "C"
    elif x>= 40:
        return "D"
    elif x>= 33:
        return "E"
    else:
        return "fail"
df["grade"] = df.percentage.apply(grade)
df.head()
```

```python
df.grade.value_counts()
```

```python
dic = {"fail":7,
       "E":6,
       "D":5,
       "C":4,
       "B":3,
       "A":2,
       "A+":1}
df["grade1"] = df.grade.map(dic)
df.head()
```

```python
# Find Rules with Deep learning
# input + output = Rules
# X  + Y = Rules(Grade)
```

```python
df1 = df[["math","phy","bio","chem","eng","grade"]]
df1 = df1.sample(frac=1) #shuffle data for better learning
df1.head()
```

```python
x = df1.loc[:,"math":"eng"]
x.head()
```

```python
y = df1[["grade"]]
y.head()
```

# Apply vectorization on Y
### One-hot-encoding

- A B C 
- A 1 0 0 
- B 0 1 0 
- C 0 0 1

```python
y = pd.get_dummies(y)
y.tail()
```

```python
y.drop_duplicates()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD
```

# Create your model network structure

- first-layer (input_shape = X.no_of_columns)
- last-layers (output=Y.unique_labels(unique classes))
    - activation = according to your problems (4.6)

```python
model = Sequential()

model.add(Dense(35, activation="relu",input_shape=(5,)))
model.add(Dense(40,activation="relu"))
model.add(Dense(35,activation="relu"))
model.add(Dense(7, activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(optimizer=SGD(lr=0.002),
             loss="categorical_crossentropy",
             metrics=["acc"])
```

```python
#training
model.fit(x,y,epochs=30)
```

```python
#testing set
df2 = pd.DataFrame(np.random.randint(20,100, size=(20,5)),columns=["math", "phy","bio", "chem", "eng"])
df2.head()
```

# Computer Vision problem
- load image from directory
- apply Auto labels based on your folder names (done by ImageDataGenerator)

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_data_gen = ImageDataGenerator(rescale= 1.0/255) #normalization
```

```python
train_data = train_data_gen.flow_from_directory("F:/study/AI/Q2 lectures/datasets/cat, dog data set/train",
                                               target_size=(150,150),
                                               batch_size=20,
                                               class_mode="binary")
```

```python
train_data.class_indices
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam, SGD
```

150 X 150 width X height resolution is no of pixels
150 X 150 * 3 no of pixels * RGB for each pixel --> 2D -> 1D

```python
model = Sequential()
model.add(Flatten()) #--> 2D or 3D to 1D
model.add(Dense(80, activation="relu"))
#model.add(Dense(30, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
```

```python
model.compile(loss="binary_crossentropy",
             optimizer="adam",
             metrics=["acc"])
```

```python
#training
hist = model.fit(train_data, epochs=30)
```

```python
model.evaluate(train_data)
```
