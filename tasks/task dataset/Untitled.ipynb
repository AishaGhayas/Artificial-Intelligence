##### PDSI index generally ranges from -6 to +6, with negative values denoting dry spells and positive values indicating wet spells.  

* PDSI values 0 to -.5 =normal
* -0.5 to -1.0 = incipient drought 
* -1.0 to -2.0 = mild drought 
* -2.0to -3.0 = moderate drought 
* -3.0 to -4.0 = severe drought
* greater than -4.0 = extreme drought.  

- Similar adjectives are attached to positive values of wet spells.

```python
hurricane_data = tabs[["East","North","WaterTemp"]].agg(["mean","var"])
hurricane_data = hurricane_data.unstack()
```

```python
buoy1_data = buoy1[["East","North","WaterTemp"]].agg(["mean","var"])
buoy1_data = buoy1_data.unstack()
```

```python
buoy2_data = buoy2[["East","North","WaterTemp"]].agg(["mean","var"])
buoy2_data = buoy2_data.unstack()
```

```python
comparison = pd.concat([hurricane_data,buoy1_data,buoy2_data],keys=["hurricane_data","buoy1_data","buoy2_data"])
comparison.unstack(level=1)
```

expected value and variance has been calculated for all the three datasets.
In statistics, the mean, or more technically the arithmetic mean or sample mean, can be estimated from a sample of examples. mean represents the expected value for all these variables.
In probability, the variance of some random variable X is a measure of how much values in the distribution vary on average with respect to the mean.The variance is denoted as the function Var() on the variable.

In this table mean shows expected value for specific variable and variance is the expectation of the squared deviation of a random variable from its mean. For example water temperature for hurricane dataset buoy is 27°C and this expected value can vary by 1.3°C.
